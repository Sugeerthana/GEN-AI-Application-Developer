```python
import matplotlib.pyplot as plt
# Define the neural network architecture
layers = {
   "Input Layer": [
       "Transaction Amount",
       "Transaction Time",
       "Merchant Category",
       "Customer Age",
       "Account Balance",
       "No. of Transactions Today"
   ],
   "Hidden Layer 1": ["Neuron 1", "Neuron 2", "Neuron 3", "Neuron 4"],
   "Hidden Layer 2": ["Neuron 1", "Neuron 2"],
   "Output Layer": ["Fraud (0/1)"]
}
# Plot setup
fig, ax = plt.subplots(figsize=(10, 6))
ax.axis("off")
x_positions = [0.1, 0.4, 0.7, 0.9]
y_spacing = 0.12
# Draw layers
for i, (layer_name, neurons) in enumerate(layers.items()):
   x = x_positions[i]
   y_start = 0.5 + (len(neurons) - 1) * y_spacing / 2
   for j, neuron in enumerate(neurons):
       y = y_start - j * y_spacing
       ax.text(x, y, neuron, ha='center', va='center',
               bbox=dict(boxstyle="round", ec="black"))
# Draw connections
def connect_layers(x1, y1_list, x2, y2_list):
   for y1 in y1_list:
       for y2 in y2_list:
           ax.plot([x1, x2], [y1, y2], 'k-', linewidth=0.5)
layer_y_positions = []
for neurons in layers.values():
   y_start = 0.5 + (len(neurons) - 1) * y_spacing / 2
   layer_y_positions.append(
       [y_start - i * y_spacing for i in range(len(neurons))]
   )
for i in range(len(layer_y_positions) - 1):
   connect_layers(
       x_positions[i],
       layer_y_positions[i],
       x_positions[i + 1],
       layer_y_positions[i + 1]
   )
# Layer titles
for i, layer_name in enumerate(layers.keys()):
   ax.text(x_positions[i], 0.9, layer_name, ha='center', fontsize=10, weight='bold')
plt.title("Neural Network Architecture for Credit Card Fraud Detection")
plt.show()
```
