```python
import matplotlib.pyplot as plt

# Neural Network Architecture
layers = {
   "Input\n(Transaction Features)": 6,
   "Hidden Layer 1": 5,
   "Hidden Layer 2": 4,
   "Output\n(Fraud / Non-Fraud)": 2
}

x_positions = range(len(layers))
neuron_positions = []

plt.figure(figsize=(10, 5))

# Draw neurons
for x, (layer_name, neuron_count) in zip(x_positions, layers.items()):
   y_positions = list(range(neuron_count))
   neuron_positions.append((x, y_positions))

   for y in y_positions:
       plt.scatter(x, y, s=500)

   plt.text(x, max(y_positions) + 0.8, layer_name, ha='center')

# Draw connections
for i in range(len(neuron_positions) - 1):
   x1, y1s = neuron_positions[i]
   x2, y2s = neuron_positions[i + 1]

   for y1 in y1s:
       for y2 in y2s:
           plt.plot([x1, x2], [y1, y2])

plt.axis('off')
plt.title("Neural Network Architecture for Credit Card Fraud Detection")
plt.show()
```
