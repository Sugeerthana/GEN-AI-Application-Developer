```python
import numpy as np

import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

area = np.array([800, 1000, 1200, 1500, 1800, 2000])

bedrooms = np.array([1, 2, 2, 3, 3, 4])

price = np.array([40, 50, 65, 80, 95, 120])

area_grid, bedroom_grid = np.meshgrid(

    np.linspace(area.min(), area.max(), 20),

    np.linspace(bedrooms.min(), bedrooms.max(), 20)

)

price_grid = 0.05 * area_grid + 10 * bedroom_grid - 20

fig = plt.figure(figsize=(9, 6))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(area, bedrooms, price)

ax.plot_surface(area_grid, bedroom_grid, price_grid, alpha=0.6)

ax.set_xlabel("Area (sq ft)", labelpad=10)

ax.set_ylabel("Number of Bedrooms", labelpad=10)

ax.set_zlabel("House Price", labelpad=10)

ax.set_title("3D Regression Model for House Price Prediction")

plt.tight_layout()   #  FIX

plt.show()

```
